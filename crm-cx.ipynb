Newtonian Potential
</h1>

<p style="

line-height:1.8;
">

- The Newtonian potential measures how much every point in a system influences every other point.
- Instead of modelling gravity, we model customer influence.
- Each customer contributes to the influence of every other customer.
- Customers who are more similar exert greater influence.
</p>

<h2 style="
color:#D97706;
font-size:42px;
">
Mathematical Background
</h2>

Classically,

$$
U(x)=
\int_{\Omega}
\frac{f(y)}
{|x-y|^{n-2}}
dy
$$

where

- x = observation point
- y = source point
- f(y) = source strength

In customer analytics,

- x = customer of interest
- y = every other customer
- f(y) = purchase score
