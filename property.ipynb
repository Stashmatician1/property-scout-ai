# Heat Kernel Modelling

---

## Mathematical Background

The Heat Kernel is defined as

$$
K(x,y)
=
e^{-\frac{d^2}{2\sigma^2}}
$$

where

- $d$ is the Euclidean distance from the centre.
- $\sigma$ controls the spread of influence.
- $K(x,y)$ is the resulting influence value.

## Interpretation

Properties closer to the centre have larger values of $K(x,y)$, while properties farther away experience a rapidly decreasing influence.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,4,500)
sigma = 1

y = np.exp(-(x**2)/(2*sigma**2))

plt.plot(x,y,linewidth=3)
plt.title("Heat Kernel")
plt.xlabel("Distance")
plt.ylabel("Influence")
plt.grid(True)
````

````python
X,Y = np.meshgrid(
    np.linspace(-2,2,250),
    np.linspace(-2,2,250)
)

sigma = 0.4
Z = np.exp(-((X)**2+(Y)**2)/(2*sigma**2))
plt.imshow(Z,cmap="turbo")
plt.colorbar()
````
